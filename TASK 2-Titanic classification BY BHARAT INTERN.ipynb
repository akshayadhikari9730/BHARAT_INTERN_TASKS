```python
#Importing python libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Next is importing the dataset
df=pd.read_csv("Titanic-Dataset.csv")
df
```

```python
# to see the description of data
df.info()
```

```python
# to see null values
df.isnull().sum()
```

```python
# Now to apply feature engineering
a=df['Age'].mean()
df['Age'].fillna(a,inplace=True)
```

```python
df['Age'].isnull().sum()
```

```python
# Now to preprocess the data
df.isnull().sum()
```

```python
df
```

```python
# to see unique values in "Embarked"column
df['Embarked'].unique()
```

```python
#To see count of values of Survived
df['Survived'].value_counts()
```

```python
# to see bar distribution of 
sns.barplot(x='Pclass', y='Survived', data=df)
```

```python
# now to drop column cabin as it is not needed
df.drop(['Cabin'],axis=1)
```

```python
# Now to apply encoding on few columns
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
df['Embarked']=le.fit_transform(df['Embarked'])
```

```python
df       #( 1 denotes male and 0 denotes female)
```

```python
grid = sns.FacetGrid(df, col='Survived', row='Pclass', aspect=1.6)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend();
```

```python
# Now To extract independent and dependent variables
X = df.drop(['Survived', 'Name', 'Ticket', 'Cabin','PassengerId'], axis=1)
Y = df[['Survived']]
```

```python
X
```

```python
Y
```

```python
#Now Split the data in training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#feature Scaling  
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
X_train= st_x.fit_transform(X_train)    
X_test= st_x.transform(X_test)  
```

# LOGISTIC REGRESSION

```python
# Now to apply models
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
lr_pred=lr.predict(X_test)
lr_pred
```

```python
# to check classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test,lr_pred))
```

# DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

```python
dt_pred=dt.predict(X_test)
dt_pred
```

```python
# to check classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test,dt_pred))
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=1000)
rf.fit(X_train,Y_train)
```

```python
rf_pred=rf.predict(X_test)
rf_pred
```

```python
# to check classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test,rf_pred))
```

# XGBOOST

```python
from xgboost import XGBClassifier
xg = XGBClassifier(n_estimators=1000)
xg.fit(X_train,Y_train)
```

```python
xg_pred=xg.predict(X_test)
xg_pred
```

```python
# to check classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test,xg_pred))
```

```python
#Let compare their accuracies
from sklearn.metrics import accuracy_score
lr_acc=accuracy_score(Y_test,lr_pred)
dt_acc=accuracy_score(Y_test,dt_pred)
rf_acc=accuracy_score(Y_test,rf_pred)
xg_acc=accuracy_score(Y_test,xg_pred)
print("Accuracy of logistic regression",lr_acc)
print("Accuracy of decision tree",dt_acc)
print("Accuracy of Random Forest",rf_acc)
print("Accuracy of Xgboost classifier",xg_acc)
```

```python
models=["logistic regression","decision tree","random forest","XGBoost"]
accuracies=[lr_acc,dt_acc,rf_acc,xg_acc]
plt.bar(models,accuracies, color=['red','green','blue','yellow'])
plt.xlabel("ML Models")
plt.ylabel("Accuracy")
plt.show()
```

```python
# Out of all the Algorithms, Random Forest works better with a nice accuracy followed by Xgboost
```
